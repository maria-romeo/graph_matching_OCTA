```python
import os
classes = ['baseline', 'pygm_classic', 'pygm_neural']
dropout_values = [25, 50, 75, 100]
for dropout_value in dropout_values:
    print('-------------------------')
    print(f'Dropout {dropout_value}%')
    for type_class in classes:
        

        if type_class == 'baseline':
            models = ['baseline']
        elif type_class == 'pygm_classic':
            models = ['ipfp', 'rrwm', 'sm']
        else:
            models = ['cie', 'ngm', 'pca_gm', 'ipca_gm']
        print('-------------------------')
        print(type_class)
        
        for model in models:
            print(model)
            dir = f'./GM_results_random_{dropout_value}_drop'
            metrics_file = os.path.join(dir, type_class, f'{model}_avge_evaluation_metrics.txt')
            with open(metrics_file, 'r') as f:
                lines = f.readlines()
                for line in lines:
                    if 'F1' in line:
                        print(line)
```

```python
import pandas as pd

for dropout_value in dropout_values:
    for model in ['ipfp', 'rrwm', 'sm']:
        file_path = f'GM_results_random_{dropout_value}_drop/pygm_classic/{model}_evaluation_metrics.csv'
        data = pd.read_csv(file_path)

        # Calculate the mean and standard deviation for each metric
        metrics = ['accuracy', 'precision', 'recall', 'f1']
        mean_values = data[metrics].mean()
        std_values = data[metrics].std()

        # Format the output strings
        mean_string = (f"Mean Test Accuracy: {mean_values['accuracy']}, "
                    f"Mean Test Precision: {mean_values['precision']}, "
                    f"Mean Test Recall: {mean_values['recall']}, "
                    f"Mean Test F1 Score: {mean_values['f1']}")

        std_string = (f"Std Test Accuracy: {std_values['accuracy']}, "
                    f"Std Test Precision: {std_values['precision']}, "
                    f"Std Test Recall: {std_values['recall']}, "
                    f"Std Test F1 Score: {std_values['f1']}")

        # Save to a text file
        output_file_path = f'GM_results_random_{dropout_value}_drop/pygm_classic/{model}_avge_evaluation_metrics.txt'

        with open(output_file_path, 'w') as file:
            file.write(mean_string + '\n')
            file.write(std_string)

        print(f"Metrics saved to {output_file_path}")
```
